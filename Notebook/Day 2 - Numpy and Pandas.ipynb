## DataFrame
DataFrame adalah struktur data 2 dimensi yang berbentuk tabular (mempunyai baris dan kolom). Setiap kolom pada Data Frame merupakan objek dari Series, dan baris terdiri dari elemen yang ada pada Series.

```python
# create dataframe dari dict
dict_staff = {
    'div':[rd.choice(['backend','frontend','ui/ux']) for i in range(50)],
    'jk':[rd.choice(['L','P']) for i in range(50)],
    'age':[rd.choice(range(23,37)) for i in range(50)],
    'rating':[rd.uniform(0.0, 10.0) for i in range(50)],
    'branch':[rd.choice(['Jkt','Bdg','Yk','Sby','Dps']) for i in range(50)]
}
```

```python
df_staff = pd.DataFrame(dict_staff)
```

```python
df_staff.head()
```

```python
df_staff.describe()
```

```python
df_staff.info()
```

```python
df_staff.columns
```

```python
df_staff.shape
```

## Filtering 

```python
df_staff[df_staff.jk == 'P']
```

```python
df_staff[staff_pr & staff_yk]
```

```python
df_staff.jk.value_counts()
```

```python
df_staff[staff_yk].count()
```

## Map & Apply

```python
df_staff.head()
```

```python
df_staff['jen_kel'] = df_staff.jk.map({'L':'Laki-laki','P':'Perempuan'})
```

```python
df_staff.head()
```

```python
def rating_up(x):
    return x * 10
```

```python
df_staff['rating_up'] = df_staff.rating.apply(rating_up)
```

```python
df_staff.head()
```

## Scaling 

```python
# fungsi normalisasi
def normalisasi(x):
    a = x - x.min()
    b = x.max() - x.min()
    return a/b
```

```python
df_staff['age_norm'] = df_staff[['age']].apply(normalisasi)
```

```python
df_staff.head()
```

## Sorting

```python
df_staff.sort_values('rating', ascending=False).head()
```

```python
df_staff.sort_values(['age','rating'], ascending=[True, False])
```

## Pivot Tabel

```python
df_staff.pivot_table(index='jk', columns='branch', values='rating', aggfunc='mean')
```
