```python
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
```

```python
dict1 = {
    'brand':[rd.choice(['A', 'B']) for i in range(100)],
    'price':[rd.choice([15200, 21000, 27400, 30500, 23000]) for i in range(100)],
    'qty':[rd.choice(range(1,20)) for i in range(100)],
    'month':[rd.choice(['2020-10','2020-09','2020-08','2020-07']) for i in range(100)]
}
```

```python
df1 = pd.DataFrame(dict1)
```

```python
df1.head()
```

```python
df1.tail()
```

```python
# Menambahkan kolom baru di DF
df1['total'] = df1.price * df1.qty
```

```python
df1.head()
```

```python
order = df1.pivot_table(index='month', columns='brand', values='total', aggfunc='sum')
```

```python
order
```

```python
order.index
```

```python
order.A
```

```python
order.B
```

## Line Plot 

```python
x = order.index
y1 = order.A
y2 = order.B
```

```python
fig = plt.figure(figsize=(10,4))
ax = plt.axes()
ax.plot(x,y1,'r-o', label='Brand A')
ax.plot(x,y2,'g-v', label='Brand B')
ax.legend(loc='lower right')

plt.xlabel('Order Month')
plt.ylabel('Order Total')
plt.title('Monthly Report 2020', fontsize=15, color='blue')

plt.grid()
plt.show()
```

## Scatter Plot 

```python
dict2 = {
    'uts':[rd.choice(range(50,100)) for i in range(50)],
    'uas':np.linspace(50,100, 50)
}
```

```python
df2 = pd.DataFrame(dict2)
```

```python
df2.head()
```

```python
a = df2.uts
b = df2.uas
```

```python
fig = plt.figure()
plt.scatter(a, b)
plt.grid()
plt.xlabel('Nilai UTS')
plt.ylabel('Nilai UAS')
plt.title('Grafik persebaran Nilai UTS dan UAS')

plt.show()
```

## Bar Chart 

```python
fig = plt.figure(figsize=(10,4))
ax = plt.axes()
width = 0.3

x1 = np.arange(len(x))
ax.bar(x1-width/2, y1, width=width, label='Brand A')
ax.bar(x1+width/2, y2, width=width, label='Brand B')

ax.set_xticks(x1)
ax.set_xticklabels(x)

plt.legend()
plt.show()
```

## Histogram

```python
dict3 = {
    'age':[rd.choice(range(20,45)) for i in range(200)]
}
```

```python
df3 = pd.DataFrame(dict3)
```

```python
df3.head()
```

```python
df3['age_bin'] = pd.cut(x=df3['age'], bins=[20,25,35,45], labels=['J1','J2','S'])
```

```python
df3.tail()
```

```python
df3.age_bin.value_counts()
```

```python
fig = plt.figure()
plt.hist(df3.age, bins=20)

plt.show()
```

## Subplot 

```python
fig, ax = plt.subplots(2,3, figsize=(12,6))

ax[0,1].hist(df3.age, bins=20)

ax[1,2].scatter(a, b)
plt.grid()

ax[1,0].plot(x,y1,'r-o', label='Brand A')
ax[1,0].plot(x,y2,'g-v', label='Brand B')

plt.show()
```

## 3D Plot

```python
from mpl_toolkits import mplot3d
```

```python
p = np.linspace(-3,3,50)
q = np.linspace(-3,3,50)
```

```python
p
```

```python
q
```

```python
P, Q = np.meshgrid(p, q)
```

```python
P
```

```python
Q
```

```python
R = np.sqrt(P**2 + Q**2 + 50)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

ax.plot_surface(P, Q, R, cmap = 'winter')
ax.view_init(elev=15)
ax.set(xlabel='P', ylabel='Q', zlabel='R')

plt.show()
```
